```python
*  all the attributes have a outliers means 8 out-off 11 .these lower level and upper level outliers
are replaced with 5th(.05) and 95th(.95) percentile respectively.
```

```python
* numpy.clip()function is used to clip(limit)the values in an array.

SYNTAX=numpy.clip(a,a_min,a_max,out=None)
parameters,
   a :Array containing elements to clip
a_min:Minimum value
a_Max:Maximum value
if clipping is not performed on the lower interval edge.not more than one of a_min and a_max may be None  
```

```python
data['normalized-losses']=data['normalized-losses'].clip(lower=data['normalized-losses'].quantile(.05),upper=data['normalized-losses'].quantile(.95))
```

```python
data['wheel-base']=data['wheel-base'].clip(lower=data['wheel-base'].quantile(.05),upper=data['wheel-base'].quantile(.95))
```

```python
data['length']=data['length'].clip(lower=data['length'].quantile(.05),upper=data['length'].quantile(.95))
```

```python
data['width']=data['width'].clip(lower=data['width'].quantile(.05),upper=data['width'].quantile(.95))
```

```python
data['engine-size']=data['engine-size'].clip(lower=data['engine-size'].quantile(.05),upper=data['engine-size'].quantile(.95))
```

```python
data['city-mpg']=data['city-mpg'].clip(lower=data['city-mpg'].quantile(.05),upper=data['city-mpg'].quantile(.95))
```

```python
data['highway-mpg']=data['highway-mpg'].clip(lower=data['highway-mpg'].quantile(.05),upper=data['highway-mpg'].quantile(.95))
```

```python
data['compression-ratio']=data['compression-ratio'].clip(lower=data['compression-ratio'].quantile(.05),upper=data['compression-ratio'].quantile(.95))
```

# Data PreProcessing

```python
* This data set has 15 categorical variable and most of them have more than 2 categories.
  we can not run a regression model on text data,so in order to deal with the challenge lets learn about 
  label encoding.label encoding is the process of converting categorical (text) data into numerical data

in this example we will be working with the variable 'body-styyle' which has 5 category namely-
sedan          96
hatchback      70
wagon          25
hardtop         8
convertible     6
```

```python
data['body-style'].value_counts()
```

```python
# label encoding 
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder 
labelencoder=LabelEncoder()
data['body-style']=labelencoder.fit_transform(data['body-style'])
```
